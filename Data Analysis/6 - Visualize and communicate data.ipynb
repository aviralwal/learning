# Share data through visualization

Once collection, preprocessing and analysis part for data is completed, the next step comes up as communication of the finding to the required stakeholders.

## Understanding data visualization

Data visualization is the graphic representation and presentation of data. An offline example is map. Map is visualization of geological data in order to communicate data to be understood in a simpler form.

For data visualization, generally the time is divided in two ways - 
1. looking at visuals in order to understand and draw conclusions about data
2. creating visuals from raw data to tell a story.

Quick rule for creating a visualization is that audience should know exactly what they're looking at within the first five seconds of seeing it.



Types of visualizations - 
* Bar graphs - use size contrast to compare two or more values. Bar graphs are a great way to clarify trends. 
* Line graphs - help understand shifts or changes in data. 
* Pie charts show how much each part of something makes up the whole.
* Maps - help organize data geographically.
* Histogram - shows how often data values fall into certain ranges.
* Correlation - show relationships among data
* Heatmaps - Use color to compare different categories in data. Used to show relations.
* Scatter plots - show relationships between different variables.
* Distribution graph - displays the spread of various outcomes in a dataset. 

One way visualizations can be manipulated is with scaling and proportions.

One of biggest considerations when creating a data visualization is where it is needed for audience to focus. As a general rule, we should show data only that is required for understanding the findings unless it is not misleading. We can show change over time through visualizations. We can show only the time period relevant to analysis. Another way we can show data is how it is distributed. Histogram can be used for this.

Causation or a cause-effect relationship occurs when an action directly leads to an outcome. We can show correlation in data using visualization but it is important that it should not be misleading that it seems that there is causation in data. Data variables may be correlated in nature but may look like one causes other which is wrong.

Visualizations can be static or dynamic. Static visualizations do not change over time unless they're edited. Dynamic visualizations are interactive or change over time. User can interact with these visualizations. It is important to choose when to use static and when to use dynamic visualizations. Dynamic are more innteractive but give more control to user thereby reduce the focus on the story that is being told.

## The art of visualization

Some of the elements of art to take care of while performing visualization are - 
* Line - Lines and visualizations can be curved or straight, thick or thin, vertical, horizontal, or diagonal.
* Shape - Shapes and visualizations should always be two-dimensional. 3-D Visualizations can make it complicated to understand. Shapes with symmetry are usually more familiar to people, so there's less work for the audience to do when viewing symmetrical data
* Color
* Space - Space is the area between, around and in the objects. There should always be space in data visualizations, just not too much or too little. This can help in defining boundaries.
* Movement

There are nine basic principles of design that data analysts should think about when building their visualizations. These are as below - 


1. Balance: The design of a data visualization is balanced when the key visual elements, like color and shape, are distributed evenly.
2. Emphasis: Your data visualization should have a focal point, so that your audience knows where to concentrate.
3. Movement: Movement can refer to the path the viewer’s eye travels as they look at a data visualization, or literal movement created by animations.
4. Pattern: You can use similar shapes and colors to create patterns in your data visualization.
5. Repetition: Repeating chart types, shapes, or colors adds to the effectiveness of your visualization.
6. Proportion: Proportion is another way that you can demonstrate the importance of certain data. Using various colors and sizes helps demonstrate that you are calling attention to a specific visual over others.
7. Rhythm: This refers to creating a sense of movement or flow in your visualization. Rhythm is closely tied to the movement principle.
8. Variety: Your visualizations should have some variety in the chart types, lines, shapes, colors, and values you use. Variety keeps the audience engaged. But it is good to find balance since too much variety can confuse people.
9. Unity: The last principle is unity. This means that your final data visualization should be cohesive. If the visual is disjointed or not well organized, it will be confusing and overwhelming.

Choosing the right visualization for data findings can often come down to one question, "Which one will make it easiest for the user to understand the point you're trying to make?"

**Data composition** is achieved by combining the individual parts of a visualization and displaying them together as a whole. Stack bars, donuts, stacked areas, pie charts and tree maps can do all this. To show relationships in your data, you might want to use scatterplot and bubble charts, column/line charts and heatmaps. For comparing data over time, line graphs could be effective like bar graphs or stacked bar graphs or area charts. To compare distinct objects, ordered bar chart, group bar chart and ordered column chare are useful.

Elements of effective visuals - 
* Clear meaning - good visualizations clearly communicate their intended insight.
* Sophisticated use of contrast -  helps separate the most important data from the rest using visual context that our brains naturally look for.
* Refined execution - Visuals with refined execution include deep attention to detail, using visual elements like lines, shapes, colors, value, space and movement.

**Design thinking** is a process used to solve complex problems in a user-centric way. If you use design thinking when planning and creating your data viz, you'll be making decisions based on the needs of the people who will be viewing them. This way your audience will be engaged and enlightened by how you visualize your findings.

When applying design thinking in data visualization, we can do that using the 5 phases below - 
* Empathize - Think about the emotions and needs of the target audience of your data viz, whether it's stakeholders, team members or the general public.
* Define - helps find your audiences needs, their problems, and your insights. Can use this phase to think about which data to show in your visualization.
* Ideate - start to generate your data viz ideas. You'll use all of your findings from the empathize and define phases to brainstorm potential data viz solutions.
* Prototype - start putting your charts, dashboards or other visualizations together.
* Test - You could test your visualizations by showing them to team members before presenting them to stakeholders.

One of the easiest ways to highlight key data in your data viz, is through headlines. A **headline** is a line of words printed in large letters at the top of the visualization to communicate what data is being presented. Then, explain your data viz even further with a subtitle. A **subtitle** supports the headline by adding more context and description. A **legend** or key identifies the meaning of various elements in a data visualization and can be used as an alternative to labeling data directly. A **label** in a visualization identifies data in relation to other data. An **annotation** briefly explains data or helps focus the audience on a particular aspect of the data in a visualization. 

Ways to make data visualizations more accessible are -
