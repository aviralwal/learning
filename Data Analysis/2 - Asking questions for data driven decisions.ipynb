# Asking questions for data driven decisions

The first step in data analysis is to ask questions in order to make data driven decisions. These are Business Challenge/Objective/Question, i.e., what is it that is exactly required to be done and the expectation of the stakeholders is correctly understood. Focus on the problem. One way to understand what exactly is the root question is by asking the 5 whys. For answer to each why, ask a new why. By the 5th why, you will reach to the exact root question or will be near to it.

To get to the root of problems and find practical solutions, the first and most important step is understanding it.

## Asking effective questions

Just asking questions is not important, it is important to ask good questions that can help in the process of data analysis. A measure of good question can be one which follow SMART characterestics -
1. **S**pecific
2. **M**easurable. 
3. **A**ction-oriented
4. **R**elevant
5. **T**ime-bound

Question should also be fair. They should not be misleading and should make any assumptions

## Structured thinking

Defining a structure of problem is important before jumping into the problem since it helps to guide and outline how can the problem be approached. This is known as problem domain. **Problem domain** is defined as the specific area of analysis that excompasses every activity affecting or affected by the problem. Understanding problem domain is important before doing anything.

Understanding the problem domain can be done through structured thinking. **Structured thinking** is the process of recognizing the current problem or situation, organizing available information, revealing gaps and opportunities, and identifying the options. In other words, it's a way of being super prepared. It's having a clear list of what you are expected to deliver, a timeline for major tasks and activities, and checkpoints so the team knows you're making progress.

The starting place for structured thinking is the problem domain. Once you know the specific area of analysis, you can set your base and lay out all your requirements and hypotheses before you start investigating.

Another way that you can practice structured thinking and avoid mistakes is by using a scope of work. A scope of work or SOW is an agreed- upon outline of the work you're going to perform on a project. A data analyst's scope of work will include those the above items, but also focus on things like data preparation, validation, analysis of quantitative and qualitative datasets, initial results, and maybe even some visuals to really get the point across.

## Importance of context(The importance of contextualizing data, and recognizing data bias)

Data doesn't live in a vacuum, it needs context. **Context** is the condition in which something exists or happens. Figuring out what data means, is just as important as collecting it. As a data analyst, a big part of job, is putting data into context. It's also up to you, to remain objective and recognize all sides of an argument, before drawing conclusions. Conclusions can be influenced by your own conscious and subconscious biases, which are based on cultural, social and market norms among others.

Which brings us to a major limitation of data analytics. If the analysis is not objective, the conclusions can be misleading. To really understand what the data is about, we have to think through who, what, where, when, how and why. We can ask the following questions - 
1. Who - who collected the data? And what is it about? 
2. What - What does the data represent in the world, and how does it relate to other data?
3. Where - Where was the data specifically collected and is it relevant to the data rqeuired.
3. When - When, was the data collected? Data collected awhile ago may have certain limitations, given the present day situation.
4. How - What was the source of data collection? Surveys, datasets?
5. Why - Why was specific data collected? Does it contain any information that may help in the data analysis process.

The best thing for the fairness and accuracy of data, is to make sure to start with an accurate representation of the data, and collect the data in the most appropriate, and objective way.

## Communication is key

Keeping stakeholders expectations in focus is important since they hold the stake in the work. There's a good chance they'll need the work done to perform their own needs. That's why it's important to make sure the work lines up with their needs and why it is needed to communicate effectively with all of the stakeholders across the team. Stakeholders will want to discuss things like the project objective, what is needed to reach that goal, and any challenges or concerns are there. These conversations help build trust and confidence in your work.

Before communicating anything, think about who the audience is, what they already know, what they need to know and how can that be communicated effectively to them.
